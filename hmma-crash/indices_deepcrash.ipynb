## Preparing data from Yahoo (from distant past)

```python
def yahoo_scrape(symbol, start_year=1960):
    """Scrape via yahoo API to obtain data."""
    tyear, tmonth, tday = (datetime.today().year, datetime.today().month, datetime.today().day)
    final_columns = ['open', 'high', 'low', 'close', 'volume',
                     'adj_open', 'adj_high', 'adj_low', 'adj_close', 'adj_volume']
    df = pd.DataFrame(columns=final_columns)
    numtrial = 0
    try:
        while df.shape[0] == 0 and numtrial < 5:
            data_yahoo = Fetcher(symbol, [start_year, 1, 1], [tyear, tmonth, tday])
            df = data_yahoo.getHistorical()
            df.set_index('Date', inplace=True)
            df.index = df.index.astype('datetime64[ns]')
            df.dropna(inplace=True)
            df = df[(df['Volume'] > 0) | (df['High'] > df['Low'])]
            df.columns = ['open', 'high', 'low', 'close', 'adj_close', 'volume']
            adj_factor = df['adj_close'] / df['close']
            df['adj_open'] = df['open'] * adj_factor
            df['adj_high'] = df['high'] * adj_factor
            df['adj_low'] = df['low'] * adj_factor
            df['adj_volume'] = (df['volume'] / adj_factor).apply(int)
            df = df[final_columns].round(4)
            print('symbol %s has %s rows of data.' % (symbol, df.shape[0]))
            numtrial += 1
        print('Successfully download symbol %s data with lastdate on %s.' % (symbol, df.index[-1]))
    except Exception as ex:
        template = "An exception of type {0} occurred. Arguments:\n{1!r}"
        print(template.format(type(ex).__name__, ex.args), '\n')
        print('Something goes wrong when scraping symbol %s.' % (symbol))
    return df

list_fields = ['open', 'high', 'low', 'close']
adj_columns = ['adj_%s'%(field) for field in list_fields]

dict_indices = dict()
for symbol in ['^GSPC', '^IXIC', '^RUT', '^DJI', '^NYA']:
    df_symbol = yahoo_scrape(symbol, start_year=1960)
    df_symbol = np.round(df_symbol[adj_columns], 2)
    df_symbol = df_symbol.rename(columns={'adj_%s'%(field): '%s'%(field) for field in list_fields})
    dict_indices[symbol] =  df_symbol
    print(df_symbol.info())
    
dict_etf = dict()
for symbol in ['SPY', 'QQQ', 'IWM', 'DIA', 'TLT', 'VTI']:
    df_symbol = yahoo_scrape(symbol, start_year=1994)
    df_symbol = np.round(df_symbol[adj_columns], 2)
    df_symbol = df_symbol.rename(columns={'adj_%s'%(field): '%s'%(field) for field in list_fields})
    dict_etf[symbol] =  df_symbol
    print(df_symbol.info())
```

## Functions for preparing statistics

```python
def atm_pctprice(sigma, tau):
    """
    Compute percentage pricing of ATM option given annualized volatility and duration to expiry.
    (Assuming zero risk-free interest rate)
    """
    d_1 = sigma / 100 * np.sqrt(tau / 365) / 2
    d_2 = -sigma / 100 * np.sqrt(tau / 365) / 2
    call = norm.cdf(d_1) - norm.cdf(d_2)
    put = norm.cdf(-d_2) - norm.cdf(-d_1)
    return 50 * (call + put)

def log_move(df):
    """Obtain the logarithmic change of overnight and intraday movement."""
    col_logpct = 100 * np.log(df['close'] / df['close'].shift(1))
    col_c2ochg = 100 * np.log(df['close'] / df['open'])
    col_range = 100 * np.log(df['high'] / df['low'])
    df_log = pd.concat([col_logpct, col_c2ochg, col_range], axis=1, keys=['log_pctchg', 'c2o_pctchg', 'log_range'])
    return pd.concat([df, df_log], axis=1)

def chgsigma(df_log, cycle=21):
    """Obtain relevant features about geometric Bollinger Band."""
    col_chg = (df_log['close'] / df_log['close'].shift(cycle) - 1) * 100
    col_sigma = df_log['log_pctchg'].rolling(cycle).std() * np.sqrt(252)
    col_var = col_sigma * col_sigma
    col_atm = atm_pctprice(np.sqrt(col_var.ewm(span=cycle, min_periods=6).mean()), cycle * 10/7)
    col_rar = col_chg / col_atm
    df_concat = pd.concat([col_chg, col_sigma, col_rar], axis=1,
                          keys=['chg_%s' % (cycle), 'sigma_%s' % (cycle), 'RAR_%s' % (cycle)])
    return df_concat

def rolling_ddpb(df, cycle=252):
    """Obtain the rolling ."""
    col_dd = df['low'] / df['high'].rolling(cycle).max() - 1
    col_pb = np.log(df['high'] / df['low'].rolling(cycle).min())
    df_concat = pd.concat([col_dd, col_pb], axis=1, keys=['DD_%s'%(cycle), 'PB_%s'%(cycle)])
    
    return df_concat

def cyclehl(df, cycle=252):
    col_cycleh = df['high'].rolling(cycle).max()
    col_cyclel = df['low'].rolling(cycle).min()
    col_range = np.log(col_cycleh / col_cyclel)
    list_dateh = [df.index[0]]
    list_datel = [df.index[0]]
    for num in range(1, len(df.index)):
        if col_cycleh[num] == df['high'][num]:
            list_dateh.append(df.index[num])
        else:
            list_dateh.append(list_dateh[-1])
        if col_cyclel[num] == df['low'][num]:
            list_datel.append(df.index[num])
        else:
            list_datel.append(list_datel[-1])  
    df_hldates = pd.DataFrame.from_dict({'cycleh_date': list_dateh, 'cyclel_date': list_datel})
    df_hldates = df_hldates.set_index(df.index)
    
    df_concat = pd.concat([col_cycleh, col_cyclel, col_range], 
                           axis=1, keys=['cycleh_%s'%(cycle), 'cyclel_%s'%(cycle), 'log_range'])   
    df_concat = pd.concat([df_concat, df_hldates], axis=1)    
    df_concat['numdays_hi'] = df_concat.index - df_concat['cycleh_date']
    df_concat['numdays_lo'] = df_concat.index - df_concat['cyclel_date']
    
    return df_concat

```

## Indices volatility and drawdown statistics

```python
def get_statkpi(datadict, symbol, fast_cycle=21, trend_cycle=252):
    """Append statistical features on data time series."""
    df_symbol = datadict[symbol]
    log_symbol = log_move(df_symbol)
    rar_symbol = chgsigma(log_symbol, fast_cycle)
    ddpb_symbol = rolling_ddpb(df_symbol, trend_cycle)
    hldates_symbol = cyclehl(df_symbol, trend_cycle)
    stat_symbol = pd.concat([log_symbol, rar_symbol, ddpb_symbol, hldates_symbol], axis=1)
    
    return stat_symbol
    
stat_spx = get_statkpi(dict_indices, '^GSPC')
stat_dow = get_statkpi(dict_indices, '^DJI')
stat_spy = get_statkpi(dict_etf, 'SPY')
stat_dia = get_statkpi(dict_etf, 'DIA')
```

### Breaking down some crashes

```python
def hmean(numlist):
    """Compute harmonic mean of a series."""
    return np.round(len(numlist) / sum([1.0/num for num in numlist]), 4)

for cycle in [fast_cycle, mid_cycle, slow_cycle, trend_cycle]:
    stat_spx['hmean_%s'%(cycle)] = stat_spx['close'].rolling(cycle).apply(hmean)
```

```python
crashstart = stat_spx.index[-21]
crashend = stat_spx.index[-1]
spx_crashbuy = stat_spx.loc[crashstart:crashend] # Crash period
avgcost_crashbuy = hmean(spx_crashbuy['close'])
pct_crashbuy = stat_spx['close'][-1] / avgcost_crashbuy - 1
print('SPX average cost on 21-TD buying during %s to %s: %1.4f'%(crashstart.strftime('%Y-%m-%d'), 
                                                                 crashend.strftime('%Y-%m-%d'), avgcost_crashbuy))
print('SPX latest pct return for 21-TD average buying : %1.4f'%(pct_crashbuy))

for cycle in [fast_cycle, mid_cycle]:
    stat_spxunder = stat_spx[stat_spx['hmean_%s'%(cycle)] <= stat_spx['hmean_252']]
    print('\nPct of days %s-HMMA under 252-HMMA: %1.4f'%(cycle, 100 * stat_spxunder.shape[0] / (stat_spx.shape[0] - 251)))
    print('Average return on %s-TD buying under 252 HMA: %1.4f'%(cycle, stat_spxunder['pct_avg%s'%(cycle)].mean()))
    print('SD return on %s-TD buying under 252 HMA: %1.4f'%(cycle, stat_spxunder['pct_avg%s'%(cycle)].std()))
    print('rank of most recent thrust: %1.4f\n'%(percentileofscore(stat_spxunder['pct_avg%s'%(cycle)], pct_crashbuy)))
    print('Average %s-TD forward return: %1.4f'%(cycle, stat_spxunder['fwd_%s'%(cycle)].mean()))
    print('SD %s-TD forward return: %1.4f'%(cycle, stat_spxunder['fwd_%s'%(cycle)].std()))
    print('Average 1-yr forward return: %1.4f'%(stat_spxunder['fwd_252'].mean()))
    print('SD 1-yr forward return: %1.4f\n'%(stat_spxunder['fwd_252'].std()))  
    
    for pctile in [.99, .95, .85]:
        pctile_return = stat_spxunder['pct_avg%s'%(cycle)].quantile(pctile)
        print('%sth pctile backward return: %1.4f'%(int(100*pctile), pctile_return))
        stat_spxpctile = stat_spxunder[stat_spxunder['pct_avg%s'%(cycle)] >= pctile_return]
        print('Average %s-TD forward return: %1.4f'%(cycle, stat_spxpctile['fwd_%s'%(cycle)].mean()))
        print('SD %s-TD forward return: %1.4f'%(cycle, stat_spxpctile['fwd_%s'%(cycle)].std()))
        print('Average 1-yr forward return: %1.4f'%(stat_spxpctile['fwd_252'].mean()))
        print('SD 1-yr forward return: %1.4f\n'%(stat_spxpctile['fwd_252'].std()))        
        
    stat_spxabove = stat_spx[stat_spx['hmean_%s'%(cycle)] > stat_spx['hmean_252']]
    print('\nPct of days %s-HMMA above 252-HMMA: %1.4f'%(cycle, 100 * stat_spxabove.shape[0] / (stat_spx.shape[0] - 251)))
    print('Average return on %s-TD buying under 252 HMA: %1.4f'%(cycle, stat_spxabove['pct_avg%s'%(cycle)].mean()))
    print('SD return on %s-TD buying under 252 HMA: %1.4f'%(cycle, stat_spxabove['pct_avg%s'%(cycle)].std()))
    print('rank of most recent thrust: %1.4f'%(percentileofscore(stat_spxabove['pct_avg%s'%(cycle)], pct_crashbuy)))
    for pctile in [.99, .95, .75]:
        pctile_return = stat_spxabove['pct_avg%s'%(cycle)].quantile(pctile)
        print('%sth pctile backward return: %1.4f'%(int(100*pctile), pctile_return))
        stat_spxpctile = stat_spxabove[stat_spxabove['pct_avg%s'%(cycle)] >= pctile_return]
        print('Average %s-TD forward return: %1.4f'%(cycle, stat_spxpctile['fwd_%s'%(cycle)].mean()))
        print('SD %s-TD forward return: %1.4f'%(cycle, stat_spxpctile['fwd_%s'%(cycle)].std()))
        print('Average 1-yr forward return: %1.4f'%(stat_spxpctile['fwd_252'].mean()))
        print('SD 1-yr forward return: %1.4f\n'%(stat_spxpctile['fwd_252'].std()))   
```
