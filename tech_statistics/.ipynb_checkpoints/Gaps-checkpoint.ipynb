```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from datetime import date, datetime, timedelta
from yahoo_historical import Fetcher
```

## Preparing data

```python
def get_adjustyahoo(symbol, start_year=1994):
    """Scrape via yahoo API to obtain adjusted data."""
    tyear, tmonth, tday = (datetime.today().year, datetime.today().month, datetime.today().day)
    list_fields = ['open', 'high', 'low', 'close']
    final_columns = ['adj_%s'%(field) for field in list_fields]
    df = pd.DataFrame(columns=final_columns)
    numtrial = 0
    try:
        while df.shape[0] == 0 and numtrial < 5:
            data_yahoo = Fetcher(symbol, [start_year, 1, 1], [tyear, tmonth, tday])
            df = data_yahoo.getHistorical()
            df.set_index('Date', inplace=True)
            df.index = df.index.astype('datetime64[ns]')
            df.dropna(inplace=True)
            df = df[(df['Volume'] > 0) | (df['High'] > df['Low'])]
            df.columns = ['open', 'high', 'low', 'close', 'adj_close', 'volume']
            adj_factor = df['adj_close'] / df['close']
            df['adj_open'] = df['open'] * adj_factor
            df['adj_high'] = df['high'] * adj_factor
            df['adj_low'] = df['low'] * adj_factor
            df = df[final_columns].round(4)
            df = df.rename(columns={'adj_%s'%(field): '%s_%s'%(symbol, field) for field in list_fields})
            print('symbol %s has %s rows of data.' % (symbol, df.shape[0]))
            numtrial += 1
        print('Successfully download symbol %s data with lastdate on %s.' % (symbol, df.index[-1]))
    except Exception as ex:
        template = "An exception of type {0} occurred. Arguments:\n{1!r}"
        print(template.format(type(ex).__name__, ex.args), '\n')
        print('Something goes wrong when scraping symbol %s.' % (symbol))
    return df
```

```python
symbollist = ['SPY', 'IVV', 'QQQ', 'XLK', 'GLD', 'TLT', '^HSI']
  
dict_symbols = {symbol: get_adjustyahoo(symbol) for symbol in symbollist}
```

```python
# Correcting 2020-02-24 SPY data (!) with reliable IB intraday data 

dict_symbols['SPY'].loc['2020-02-24', 'SPY_high'] = 322.5160
dict_symbols['SPY'].loc['2020-02-17':'2020-03-20']
```

## Gap study

```python
def get_pricedata(symbol, df_symbol, startdate, enddate, macycle=21, fwdlen=21):
    """Obtain historical data from database."""
    df_symbol = df_symbol.loc[startdate:enddate]
    df_symbol = df_symbol.rename(columns={'%s_%s'%(symbol, field): '%s'%(field) for field in ['open', 'high', 'low', 'close']})
    df_symbol['pct'] = np.log(df_symbol['close'] / df_symbol['close'].shift(1)) 
    df_symbol['o2yl'] = np.log(df_symbol['open'] / df_symbol['low'].shift(1))
    df_symbol['h2yl'] = np.log(df_symbol['high'] / df_symbol['low'].shift(1))
    df_symbol['o2yh'] = np.log(df_symbol['open'] / df_symbol['high'].shift(1))
    df_symbol['l2yh'] = np.log(df_symbol['low'] / df_symbol['high'].shift(1))
    df_symbol['%sma'%(macycle)] = df_symbol['close'].rolling(macycle).mean().shift(1)
    df_symbol['diffma'] = np.log(df_symbol['close'].shift(1) / df_symbol['%sma'%(macycle)])
    df_symbol['%sfwd'%(fwdlen)] = np.log(df_symbol['close'].shift(-fwdlen) / df_symbol['close'])
    df_symbol['candle'] = 100 * (df_symbol['close'] - df_symbol['open']) / (df_symbol['high'] - df_symbol['low'])
    df_symbol['sigma'] = df_symbol['pct'].rolling(macycle).std()
    df_symbol['diffsig'] = df_symbol['diffma'] / df_symbol['sigma']
    df_symbol = np.round(df_symbol, 4)
    
    return df_symbol
```

### 1994-2019

```python
macycle = 63
symbollist = ['SPY', 'IVV', 'QQQ', 'XLK', 'GLD', 'TLT', '^HSI']
dict_data = {symbol: get_pricedata(symbol, df_symbol, '1990-01-01', '2020-02-21', macycle) 
             for symbol, df_symbol in dict_symbols.items()}
```

```python
display_field = ['o2yl', 'h2yl', 'diffma', '21fwd', 'candle', 'diffsig']
df_spy = dict_data['SPY'].copy()
df_spy['diffmatd'] = np.log(df_spy['close'] / df_spy['%sma'%(macycle)].shift(-1))

df_spy = df_spy.sort_values(['h2yl', 'diffma'], ascending=True)
con_gap = (df_spy['o2yl'] < -0) & (df_spy['diffma'] > 0)
df_gap = df_spy[con_gap]
print('All gaps above MA:\n %s \n'%(df_gap[display_field].describe()))

con_hgap = (df_spy['h2yl'] < -0) & (df_spy['diffma'] > 0)
df_hgap = df_spy[con_hgap]
print('All unfilled gaps above MA:\n %s \n'%(df_hgap[display_field].describe()))

con_fgap = (df_spy['o2yl'] < 0) & (df_spy['h2yl'] >= 0) & (df_spy['diffma'] > 0)
df_fgap = df_spy[con_fgap]
print('All filled gaps above MA:\n %s \n'%(df_fgap[display_field].describe()))
```

```python
con_hgapbreak = (df_spy['h2yl'] < -0) & (df_spy['diffma'] > 0) & (df_spy['diffmatd'] < 0)
df_hgapbreak = df_spy[con_hgapbreak]
print('All unfilled gaps above MA:\n %s \n'%(df_hgapbreak[display_field].describe()))
```
